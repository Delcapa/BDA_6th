```python
from konlpy.tag import Okt
from konlpy.tag import Twitter
```

```python
### 0(부정) 리뷰

twt = Twitter()

all_words_0 = [] #빈 리스트에 담기용
parts= ['Noun'] #추출하고자 하는 품사. 여기선 명사로

for i in range(len(movie_0)): #movie의 행 수 만큼 반복
    text = movie_0['document'].iloc[i] #document 열에 담긴 문장을 text에 담음
    words = twt.pos(text) #('형태소', '품사') 형태로 쪼개 words에 담음
    words_arr =[]

    for n in words:
        if n =='EOS' or n =='':continue
        words_tmp=n[0] #형태소
        part = n[1] #품사
        if not (part in parts):continue
        words_arr.append(words_tmp) #형태소만 words_arr의 빈 리스트에 추가시켜나간다

    all_words_0.extend(words_arr)

print(all_words_0)
```

```python
# 부정 리뷰 형태소 수 카운팅 3174개
all_words_df0 = pd.DataFrame({'words':all_words_0, 'count':len(all_words_0)*[1]})
all_words_df0.groupby('words').sum()
```

```python
### 1(긍정) 리뷰

twt = Twitter()

all_words_1 = [] #빈 리스트에 담기용
parts= ['Noun'] #추출하고자 하는 품사. 여기선 명사로

for i in range(len(movie_1)): #movie의 행 수 만큼 반복
    text = movie_1['document'].iloc[i] #document 열에 담긴 문장을 text에 담음
    words = twt.pos(text) #('형태소', '품사') 형태로 쪼개 words에 담음
    words_arr =[]

    for n in words:
        if n =='EOS' or n =='':continue
        words_tmp=n[0] #형태소
        part = n[1] #품사
        if not (part in parts):continue
        words_arr.append(words_tmp) #형태소만 words_arr의 빈 리스트에 추가시켜나간다

    all_words_1.extend(words_arr)

print(all_words_1)
```

#### 부정 리뷰 유사도 분석

```python
parts =['Noun']
all_words_df_r0 = pd.DataFrame()
satisfaction = []

for n in range(len(movie_0)):
    text = movie_0['document'].iloc[n]
    words = twt.pos(text)
    words_df0 = pd.DataFrame()

    for i in words:
        if i =='EOS' or i =='':continue
        word_tmp = i[0]
        part = i[1]

        if not (part in parts):continue
        words_df0[word_tmp]=[1] #형태소로 나눈 것들은 다 1로 넣어 달라는 것

    all_words_df_r0 = pd.concat([all_words_df_r0, words_df0],ignore_index =True) # 세로로 병합

```

```python
all_words_df_r0 = all_words_df_r0.fillna(0)
all_words_df_r0
```

```python
import numpy as np
cos_sim=[]
target_text_0 =all_words_df_r0.iloc[4] #코사인 유사도 분석할 주 타겟 텍스트 설정

for i in range(len(all_words_df_r0)):
    cos_text = all_words_df_r0.iloc[i]
    cos = np.dot(target_text_0, cos_text) / (np.linalg.norm(target_text_0)*np.linalg.norm(cos_text))
    cos_sim.append(cos)

all_words_df_r0['cos_sim'] = cos_sim
all_words_df_r0 #cos_sim 열이 추가됨
```

```python
top15_0 = all_words_df_r0.sort_values('cos_sim', ascending=False).head(15)
top15_0 #상위 15개의 코사인 유사도 결과
```

```python
df = pd.DataFrame()

m0=[]
c0=[]

for i in [4,214,674,1365,266,1109,1456,425,1182,987,1362,345,696,288,690]:
  m0.append(movie_0["document"].iloc[i])

c=0
sr_0 = top15_0["cos_sim"]
sr_0 = sr_0.reset_index()

while c<15: #document열에 맞게 cos_sim을 불러옴
  c0.append(sr_0["cos_sim"].iloc[c])
  c = c + 1
df["document"] = m0
df["cos_sim"] = c0

df
```
